```python
import json
from collections import Counter
import operator
import numpy as np
import string
from nltk.stem import PorterStemmer
```

```python
pl_title = json.load(open('../MODEL_1_PL_NAME_NEW/PID_PROCESSED_TITLE_LIST_PROCESSED.json'))
```

```python
Train = dict(json.load(open('../DATA_PROCESSING/PL_TRACKS_5_TRAIN.json')).items() + json.load(open('../DATA_PROCESSING/PL_TRACKS_5_TEST.json')).items())
```

```python
len(Train)
```

```python
word_tracks_list = {}
for pl in Train:
    for word in pl_title[pl]:
        try:
            word_tracks_list[word] += Train[pl]
        except:
            word_tracks_list[word] = Train[pl]
```

```python
word_tracks_freq = {}
for word in word_tracks_list:
    word_tracks_freq[word] = Counter(word_tracks_list[word])
```

```python
word_tracks_freq['old'] = dict(Counter(word_tracks_freq['old']) + Counter(word_tracks_freq['oldi']))
```

```python
word_tracks_freq['2017'] = dict(Counter(word_tracks_freq['2017']) + Counter(word_tracks_freq['17']))
```

```python
word_tracks_freq['2016'] = dict(Counter(word_tracks_freq['2016']) + Counter(word_tracks_freq['16']))
```

```python
word_tracks_freq['2015'] = dict(Counter(word_tracks_freq['2015']) + Counter(word_tracks_freq['15']))
```

```python
word_tracks_freq['2014'] = dict(Counter(word_tracks_freq['2014']) + Counter(word_tracks_freq['14']))
```

```python
word_tracks_freq['2013'] = dict(Counter(word_tracks_freq['2013']) + Counter(word_tracks_freq['13']))
```

# Process War Title for T0

```python
WAR_T0_Title = {}
with open ('WAR_T0_Title_0628_XY.txt') as f:
    for line in f:
        current = line.split()
        WAR_T0_Title[current[0]] = current[1:]
```

```python
len(WAR_T0_Title)
```

```python
R = {}
for pl in WAR_T0_Title:
    print pl
    R[pl] = {}
    for word in WAR_T0_Title[pl]:
        try:
            R[pl] = dict(Counter(R[pl])+Counter(word_tracks_freq[word]))
        except:
            print '========= no word'
            continue
```

```python
len(R)
```

```python
#delete empty
EMPTY = []
R_noempty = {}
for pl in R:
    if R[pl] != {}:
        R_noempty[pl] = R[pl]
    else:
        EMPTY.append(pl)
```

```python
len(R_noempty)
```

```python
R_sorted_raw = {}
count = 0
for pl in R_noempty:
    print count
    count += 1
    R_sorted_raw[pl] = []
    current = sorted(R_noempty[pl].items(), key=operator.itemgetter(1), reverse = True)
    for track in current[:500]:
        R_sorted_raw[pl].append(track[0])
```

```python
#delete recommanded pl not 500
NOT_ENOUGH_500 = []

R_sorted = {}
for pl in R_sorted_raw:
    if len(R_sorted_raw[pl]) == 500:
        R_sorted[pl] = R_sorted_raw[pl][:]
    else:
        NOT_ENOUGH_500.append(pl)
        print len(R_sorted_raw[pl])
```

# Fill in Universal Most Popular

```python
most_popular_2000 = []
with open('../MODEL_0_MOST_POPULAR/POPULARITY_RANKED_TRACK_2000.txt') as f:
    for line in f:
        most_popular_2000.append(line.replace('\n',''))
```

```python
NOT_ENOUGH_500_filled = {}
for pl in NOT_ENOUGH_500:
    NOT_ENOUGH_500_filled[pl] = R_sorted_raw[pl][:]
    for mp in most_popular_2000:
        if len(NOT_ENOUGH_500_filled[pl]) == 500:
            break
        if not mp in NOT_ENOUGH_500_filled[pl]:
            NOT_ENOUGH_500_filled[pl].append(mp)
```

```python
EMPTY_filled = {}
for pl in EMPTY:
    EMPTY_filled[pl] = most_popular_2000[:500]
```
