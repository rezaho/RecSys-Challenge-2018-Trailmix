```python
import numpy as np
import operator
```

```python
QQ = np.arange(0,500,1)
Xing = np.arange(0,2500,5)
```

```python
def rank(new, current, sort_base):
    layer = list(set(new) - set(current))
    layer_index = {}
    for elem in layer:
        layer_index[elem] = sort_base[elem]
        
    layer_sorted = sorted(layer_index.items(), key=operator.itemgetter(1))
    layer_final = [i[0] for i in layer_sorted]
    return current + layer_final
```

```python
def combine_new(QQ, Xing, GT_len):
    QQ_index = {}
    for elem in QQ:
        QQ_index[elem] = len(QQ_index)
    Xing_index = {}
    for elem in Xing:
        Xing_index[elem] = len(Xing_index)
        
        
    Ax = set(Xing[:GT_len])
    Bx = set(Xing[GT_len:])
    
    Aq = set(QQ[:GT_len])
    Bq = set(QQ[GT_len:])
    
    # Layer 1
    up_to_layer1 = rank(list(Ax & Aq), [], QQ_index)
    
    # Layer 2 part 1
    up_to_layer2_1 = rank(list(Ax & Bq), up_to_layer1, Xing_index)
    
    # Layer 2 part 2
    up_to_layer2_2 = rank(list(Bx & Aq), up_to_layer2_1, QQ_index)
    
    # Layer 3 part 1
    up_to_layer3_1 = rank(list(Ax), up_to_layer2_2, Xing_index)
    
    # Layer 3 part 2
    up_to_layer3_2 = rank(list(Aq), up_to_layer3_1, QQ_index)
    
    # Layer 3 part 3
    up_to_layer3_3 = rank(list(Bx & Bq), up_to_layer3_2, Xing_index)
    
    # Layer 4 part 1
    up_to_layer4_1 = rank(list(Bx), up_to_layer3_3, Xing_index)
    
    # Layer 4 part 2
    up_to_layer4_2 = rank(list(Bq), up_to_layer4_1, QQ_index)
    
    return up_to_layer4_2
```

```python
combine_new(QQ, Xing, 20)
```

```python
def combine_raw(QQ, Xing, GT_len):
    QQ_index = {}
    for elem in QQ:
        QQ_index[elem] = len(QQ_index)
    Xing_index = {}
    for elem in Xing:
        Xing_index[elem] = len(Xing_index)
        
        
    Ax = set(Xing[:GT_len])
    Bx = set(Xing[GT_len:])
    
    Aq = set(QQ[:GT_len])
    Bq = set(QQ[GT_len:])
        
    
    layer1 = list(Ax & Aq)
    #print layer1
    layer1_index = {}
    for elem in layer1:
        layer1_index[elem] = QQ_index[elem]
        
    layer1_sorted = sorted(layer1_index.items(), key=operator.itemgetter(1))
    #print layer1_sorted 
    layer1_final = [i[0] for i in layer1_sorted]
    current = layer1_final[:]
    #print layer1_final
    
    
    layer2_1 = list((Ax & Bq))
    #print layer2
    layer2_1_index = {}
    for elem in layer2_1:
        layer2_1_index[elem] = Xing_index[elem]
        
    layer2_1_sorted = sorted(layer2_1_index.items(), key=operator.itemgetter(1))
    #print layer2_sorted 
    layer2_1_final = [i[0] for i in layer2_1_sorted]
    current = current + layer2_1_final[:]
    #print layer2_1_final
    
    layer2_2 = list((Bx & Aq))
    #print layer2
    layer2_2_index = {}
    for elem in layer2_2:
        layer2_2_index[elem] = QQ_index[elem]
        
    layer2_2_sorted = sorted(layer2_2_index.items(), key=operator.itemgetter(1))
    #print layer2_sorted 
    layer2_2_final = [i[0] for i in layer2_2_sorted]
    current = current + layer2_2_final[:]
    #print layer2_2_final
    
    
    
    
    layer3_1 = list(Ax - set(current)) 
    layer3_1_index = {}
    for elem in layer3_1:
        layer3_1_index[elem] = Xing_index[elem]
        
    layer3_1_sorted = sorted(layer3_1_index.items(), key=operator.itemgetter(1))
    #print layer2_sorted 
    layer3_1_final = [i[0] for i in layer3_1_sorted]
    current = current + layer3_1_final[:]
    #print layer3_1_final
    
    
    layer3_2 = list(Aq - set(current)) 
    layer3_2_index = {}
    for elem in layer3_2:
        layer3_2_index[elem] = QQ_index[elem]
        
    layer3_2_sorted = sorted(layer3_2_index.items(), key=operator.itemgetter(1))
    #print layer2_sorted 
    layer3_2_final = [i[0] for i in layer3_2_sorted]
    current = current + layer3_2_final[:]
    #print layer3_2_final
    
    
    layer3_3 = ((Bx & Bq) - set(current)) 
    layer3_3_index = {}
    for elem in layer3_3:
        layer3_3_index[elem] = Xing_index[elem]
        
    layer3_3_sorted = sorted(layer3_3_index.items(), key=operator.itemgetter(1))
    #print layer2_sorted 
    layer3_3_final = [i[0] for i in layer3_3_sorted]
    current = current + layer3_3_final
    #print layer3_2_final
    
    layer4_1 = list(set(Bx) - set(current))
    layer4_1_index = {}
    for elem in layer4_1:
        layer4_1_index[elem] = Xing_index[elem]
        
    layer4_1_sorted = sorted(layer4_1_index.items(), key=operator.itemgetter(1))
    #print layer2_sorted 
    layer4_1_final = [i[0] for i in layer4_1_sorted]
    current = current + layer4_1_final
    #print layer4_1_final
    
    layer4_2 = list(set(Bq) - set(current))
    layer4_2_index = {}
    for elem in layer4_2:
        layer4_2_index[elem] = QQ_index[elem]
        
    layer4_2_sorted = sorted(layer4_2_index.items(), key=operator.itemgetter(1))
    #print layer2_sorted 
    layer4_2_final = [i[0] for i in layer4_2_sorted]
    current = current + layer4_2_final
    #print layer4_2_final
    
    return current
```

```python
combine_raw(QQ, Xing, 20) == combine_new(QQ, Xing, 20)
```
